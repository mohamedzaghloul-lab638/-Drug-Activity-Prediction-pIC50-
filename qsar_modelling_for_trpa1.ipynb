```python
target=new_client.target
res=target.search('TRPA1')
for r in res :
  print(r['target_chembl_id'])
  print(r['organism'])
  print(r['pref_name'])
  print(r['target_type'])

df=pd.DataFrame(list(res))
print(len(df))
df
```

```python
# search for all active compounds against TRPA1
activity = new_client.activity
results=activity.filter (
    target_chembl_id ='CHEMBL6007',
    standard_type='IC50' ,
    relation='=' ,
    assay_type="B" ) .only([
        'canonical_smiles'
        'standard_value'
        'standard_units'
        'molecule_chembl_id'
        'assay_chembl_id'


    ])

df=pd.DataFrame(list(results))
df
```

```python
print(len(results))
```

```python
df_leads=df.dropna(subset=['molecule_chembl_id','standard_value','canonical_smiles']).drop(columns=['assay_chembl_id', 'standard_units','units','value'])
print(len(df_leads))
df_leads
```

```python
# ic50 to pic50
import numpy as np
df_leads['standard_value'] = pd.to_numeric(df_leads['standard_value'], errors='coerce')
df_leads['PIC50'] = -np.log10(df_leads['standard_value'] * 1e-9)
df_leads
```

```python
# add feautres to smiles
! pip install rdkit
```

```python
from rdkit import Chem
from rdkit.Chem import Descriptors , rdMolDescriptors

def get_feautres(smiles):
  mol=Chem.MolFromSmiles(smiles)
  if mol is None :
     return None
  return {
       'TPSA' :Descriptors.TPSA(mol),
       'Logp' : Descriptors.MolLogP(mol),
       'MolWt' : Descriptors.MolWt(mol),
       'HBA' : rdMolDescriptors.CalcNumHBA(mol),
       'HBD' : rdMolDescriptors.CalcNumHBD(mol),
       'RotBonds' : Descriptors.NumRotatableBonds(mol)
   }

df_features = df_leads['canonical_smiles'].apply (get_feautres).apply(pd.Series) # here we copy features to our smiles
# concatenation of PIC50 with features df (smiles + features)
df_model=pd.concat([df_features , df_leads[['PIC50' , 'canonical_smiles']]] , axis =1)
```

```python
df_model.dropna(inplace=True)
df_model
```

```python
from rdkit.Chem import rdFingerprintGenerator
```

```python
def smiles_to_fps(smiles, radius=2, fpsize=2048):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return None
    gen = rdFingerprintGenerator.GetMorganGenerator(radius, fpSize=fpsize)
    fp = gen.GetFingerprint(mol)
    return np.array(fp)

df_model['fingerprint'] = df_model['canonical_smiles'].apply(smiles_to_fps)
df_model = df_model.dropna(subset=['fingerprint'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor

x= np.stack(df_model['fingerprint'])
y= df_model['PIC50'].values

# splitting
x_train , x_test,y_train , y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=RandomForestRegressor(n_estimators=200,random_state=42 , max_depth=10 , n_jobs=-1 )# use all cpus
model.fit(x_train,y_train)
```

```python
# evaluation
y_pred=model.predict(x_test)
r2=r2_score(y_test , y_pred)
mse=np.sqrt(mean_squared_error(y_test , y_pred))
print(f"R2 score is : {r2:.3f}")
print(f"mse is: {mse:.3f}")
```

```python
descriptors = df_model[['MolWt', 'Logp', 'TPSA', 'HBA', 'HBD', 'RotBonds']].values
fps = np.stack(df_model['fingerprint'])

x = np.hstack([fps, descriptors])
y= df_model['PIC50'].values

# splitting
x_train , x_test,y_train , y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=RandomForestRegressor(n_estimators=200,random_state=42 , max_depth=10 , n_jobs=-1 )# use all cpus
model.fit(x_train,y_train)
```

```python
# evaluation
y_pred=model.predict(x_test)
r2=r2_score(y_test , y_pred)
mse=np.sqrt(mean_squared_error(y_test , y_pred))
print(f"R2 score is : {r2:.3f}")
print(f"mse is: {mse:.3f}")
```

```python
# using XGBoost
from xgboost import XGBRegressor
model=XGBRegressor(learning_rate=0.05 , subsample=0.8 , colsample_bytree=0.8 , random_state=42 , verbosity=0, max_depth=10)
model.fit(x_train,y_train)
```

```python
# evaluation
y_pred=model.predict(x_test)
r2=r2_score(y_test , y_pred)
mse=np.sqrt(mean_squared_error(y_test , y_pred))
print(f"R2 score is : {r2:.3f}")
print(f"mse is: {mse:.3f}")
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'n_estimators': [300, 500, 700],
    'max_depth': [4, 6, 8],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

grid = GridSearchCV(XGBRegressor(random_state=42),
                    params, cv=5, scoring='r2', n_jobs=-1)
grid.fit(x_train, y_train)
print("Best R2:", grid.best_score_)
print("Best Params:", grid.best_params_)
```

```python
# evaluation of grid
y_pred=grid.predict(x_test)
r2=r2_score(y_test , y_pred)
mse=np.sqrt(mean_squared_error(y_test , y_pred))
print(f"R2 score is : {r2:.3f}")
print(f"mse is: {mse:.3f}")
```

```python
# prediction
from rdkit import Chem
from rdkit.Chem import Descriptors , rdFingerprintGenerator
import numpy as np
import pandas as pd
```

```python
df_smiles = pd.read_excel('/content/smiles.xlsx')
df_smiles.to_csv('file.csv', index=False)
df_smiles
```

```python
def predict_features(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return None

    # Calculate descriptors
    tpsa = Descriptors.TPSA(mol)
    logp = Descriptors.MolLogP(mol)
    molwt = Descriptors.MolWt(mol)
    hba = rdMolDescriptors.CalcNumHBA(mol)
    hbd = rdMolDescriptors.CalcNumHBD(mol)
    rotbonds = Descriptors.NumRotatableBonds(mol)
    descriptors = np.array([molwt, logp, tpsa, hba, hbd, rotbonds])

    # Generate fingerprints
    gen = rdFingerprintGenerator.GetMorganGenerator(radius=2, fpSize=2048)
    fp = gen.GetFingerprint(mol)
    fingerprint = np.array(fp)

    # Combine features
    combined_features = np.hstack([fingerprint, descriptors])
    return combined_features.reshape(1, -1)
```

```python
# Load the trained model
import joblib
model = joblib.load('pic50_model.pkl')
prediction_data=[]
df_clean = df.dropna(subset=['canonical_smiles'])

for smiles in df_clean['canonical_smiles']:
    combined_features = predict_features(smiles)
    if combined_features is None:
        continue

    predicted_pic50 = model.predict(combined_features)[0]
    prediction_data.append({
        'canonical_smiles': smiles.strip(),
        'predicted_PIC50': predicted_pic50
        })

predictions_df = pd.DataFrame(prediction_data)
print(predictions_df)
```
